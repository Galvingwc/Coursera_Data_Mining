```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#import text file as dataframe
df = pd.read_csv(r'C:\Users\Galvin\Desktop\places.txt', names=['lon','lat'])
df
```

```python
#compute and predict clusters
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(df)
df
```

```python
#export dataframe as text file
df['cluster'].to_csv(r'C:\Users\Galvin\Desktop\clusters.txt', sep=" ", header=False)
```
