```python
#import libraries
import numpy as np
import pandas as pd

from sklearn.metrics import normalized_mutual_info_score, jaccard_score
```

```python
#import text file into dataframe
df_truth = pd.read_csv(r'C:\Users\Galvin\Desktop\validation\partitions.txt', sep=' ', names=['id','label'], index_col=['id'])
df_truth
```

```python
#import text files into list of dataframes
df_list = []
for i in range(1,6):
    filepath = str(i).join([r'C:\Users\Galvin\Desktop\validation\clustering_','.txt'])
    df_list.append(pd.read_csv(filepath ,sep=' ',names=['id','label'],index_col=['id']))
df_list
```

```python
#validate cluster labels
nmi = [normalized_mutual_info_score(df_truth['label'], cluster['label']) for cluster in df_list]
jac = [jaccard_score(df_truth['label'], cluster['label'], average='macro') for cluster in df_list]
```

```python
#create submission dataframe
submission = pd.DataFrame({'NMI_score':nmi,'Jac_score':jac}, columns=['NMI_score', 'Jac_score'])
submission
```

```python
#export submission dataframe as text file
submission.to_csv(r'C:\Users\Galvin\Desktop\validation\scores.txt',sep=' ',header=False,index=False)
```
